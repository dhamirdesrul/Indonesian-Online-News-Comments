## Cleaning Data

```python
RE_EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)

#removing emoji
def strip_emoji(text):
    return RE_EMOJI.sub(r'', text)

#removing symbol
def strip_symbol(text):
    return re.sub(r'[^\w\s]',' ',text)

#removing integer
def strip_integer(text):
    return re.sub(r'[\d+]', ' ', text)

#removing enter syntax
def strip_enter(text):
    return re.sub(r'[\n]', ' ', text)

#removing regex and tokenizing
def regex_tokenizer(text):
    tokenizer = RegexpTokenizer(r'\w+')
    return tokenizer.tokenize(text)
```

```python
def clean_regex(text):
    return re.sub('[\n\,\-\'\.]', ' ', text)
```

```python
def make_it_lowercase(text):
    return text.lower()
```

```python
#cleaning stopwords
def cleaning_stopwors(text):
    text = 0
    return text
```

```python
def cleaning_data(text):
    text = strip_emoji(text)
    text = strip_symbol(text)
    text = strip_integer(text)
    text = strip_enter(text)
    text = make_it_lowercase(text)
    text = regex_tokenizer(text)
    return text
```
